# Imports

```python
import os
import torch
from torch.serialization import add_safe_globals

from agent import Agent
from gym_super_mario_bros.actions import RIGHT_ONLY
from nes_py.wrappers import JoypadSpace

import gym_super_mario_bros
from wrappers import apply_wrappers
import csv
```

# Configuration

```python
PATH = 'models/old_model_1'
EPISODES_PER_LEVEL = 500
DISPLAY = False
```

# Load

```python
# Set paths and load specified model if 
LOAD = True
if LOAD:
    CKPT_PATH = os.path.join(PATH, "checkpoint.pt")
    TEST_CSV_PATH = os.path.join(PATH, "test.csv")
    TRAIN_CSV_PATH = os.path.join(PATH, "train_log.csv")

    checkpoint = torch.load(CKPT_PATH, weights_only=False)

    TRAIN_LEVELS = checkpoint['train_levels']
    TEST_LEVELS = checkpoint['test_levels']
    NUM_EVAL_EPISODES = checkpoint['num_eval_episodes']
    SKIP_FRAME = checkpoint['skip_frame']
    RESIZE = checkpoint['resize']
    FRAME_STACK = checkpoint['frame_stack']

    agent = checkpoint['agent']
```

# Test

```python
all_levels = []
for world in range(1, 9):
    for level in range(1, 5):
        all_levels.append(f"SuperMarioBros-{world}-{level}-v0")

TEST_LEVELS = []
for world in range(7, 9):
    for level in range(1, 5):
        TEST_LEVELS.append(f"SuperMarioBros-{world}-{level}-v0")

# custom = []
# for world in range(7, 9):
#     for level in range(1, 5):
#         custom.append(f"SuperMarioBros-{world}-{level}-v0")


# test_levels = [x for x in possible_levels if x not in LEVELS]
test_levels = TEST_LEVELS
# test_levels = ['SuperMarioBros-1-1-v0']
print(len(test_levels))
print(len(all_levels))
# print(len(custom))
```

### only save the runs that agent beats

```python
fields = ["level", "episode", "total_reward", "rewards_over_t"]

# for level in test_levels:
for level in all_levels:
    print(f"=== Evaluating on {level} ===")
    csv_filename = f"test_results/test_500_ep/baseline/{level}.csv"

    with open(csv_filename, mode='w', newline='') as file:
        writer = csv.DictWriter(file, fieldnames=fields)
        writer.writeheader() 
        for ep in range(EPISODES_PER_LEVEL):
            print(f"Episode {ep + 1}/{EPISODES_PER_LEVEL}")
            env = gym_super_mario_bros.make(level, render_mode='human' if DISPLAY else 'rgb', apply_api_compatibility=True)
            env = JoypadSpace(env, RIGHT_ONLY)
            env = apply_wrappers(env, SKIP_FRAME, RESIZE, FRAME_STACK)

            try:
                state, _ = env.reset()
                done = False
                total_reward = 0
                rewards = []

                while not done:
                    with torch.no_grad():
                        action = agent.choose_action(state)

                    new_state, reward, done, truncated, info = env.step(action)
                    total_reward += reward
                    state = new_state

                    rewards.append(reward)
                    if info["flag_get"] == True: # only write to the csv if mario passes this level
                        print(" ******* MARIO COMPLETED THIS LEVEL!!!!!!!")
                        writer.writerow({
                            "level": level,
                            "episode": ep,
                            "total_reward": total_reward,
                            "rewards_over_t": str(rewards) 
                        })

                # print(f"Total reward: {total_reward}")

            finally:
                env.close()

        print()
```
